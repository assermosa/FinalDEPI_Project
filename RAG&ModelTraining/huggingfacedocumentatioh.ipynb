```python
from langchain.schema import Document  # Import the Document class

import pandas as pd
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS



# Create chunks using relevant columns
chunked_docs = []
for _, row in df.iterrows():
    chunk = Document(
        page_content=(
            f"text: {row['text']}\n"
            f"source: {row['source']}\n"
        )
    )
    chunked_docs.append(chunk)
```

```python
embeddings_model = HuggingFaceEmbeddings(model_name="BAAI/bge-base-en-v1.5")
db = FAISS.from_documents(chunked_docs, embeddings_model)
```

```python
retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 5})
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B-Instruct")
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-3.2-1B-Instruct")
```

```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser


text_generation_pipeline = pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    temperature=0.2,
    max_new_tokens=300,
    device=0 ) # 0 for first GPU; change if you have multiple GPUs
llm = HuggingFacePipeline(pipeline=text_generation_pipeline)

# Set up the prompt template
prompt_template = """
<|system|>
Answer the question with the following movie details for an accurate response:

{context}

</s>
<|user|>
{question}
</s>
<|assistant|>
"""

prompt = PromptTemplate(input_variables=["context", "question"], template=prompt_template)
llm_chain = prompt | llm | StrOutputParser()
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain.chains import LLMChain

# Ensure your prompt template and llm_chain setup is correct
llm_chain = LLMChain(prompt=prompt, llm=llm, output_parser=StrOutputParser())

# Updated RAG chain for context retrieval
rag_chain = {
    "context": retriever, 
    "question": RunnablePassthrough()  # Ensuring that question passes correctly
} | llm_chain  # Passing through llm_chain directly
```

```python
question='how can i deploy a model'
response = llm_chain.invoke({"context": "answer concisely", "question": question})
final_response = response['text'].replace("<|user|>", "").replace("<|assistant|>", "").strip()
print(final_response)
```
